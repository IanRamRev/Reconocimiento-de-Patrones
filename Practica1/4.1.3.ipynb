```python
#https://likegeeks.com/es/procesar-de-imagenes-en-python/
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
import cv2

# Load an color image in grayscale
img = cv2.imread('.\\Imagenes\\lena_color_512.tif')
plt.imshow(img)
```

```python
#RGB a Escala de grises
img_gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
plt.imshow(img_gray)
```

```python
#RGB a YUV
img_yuv = cv2.cvtColor(img, cv2.COLOR_BGR2YUV)
plt.imshow(img_yuv)
```

```python
#RGB a HSV
img_hsv = cv2.cvtColor(img, cv2.COLOR_BGR2HSV)
plt.imshow(img_hsv)
```

```python
#Despliega la paleta de colores de RGB por separado

b,g,r = cv2.split(img)
plt.title("Lena_ canal rojo")
plt.imshow(r)
```

```python
plt.title("Lena_ canal azul")
plt.imshow(b)
```

```python
plt.title("Lena_ canal verde")
plt.imshow(g)
```

```python
#Despliega la paleta de colores HSV por separado
h, s, v = cv2.split(cv2.cvtColor(img, cv2.COLOR_BGR2HSV))
print(h)
print(s)
print(v)
```

```python
#SCRIPT PARA ABRIR IMAGEN CON SCIKIT
#https://scikit-image.org/docs/stable/api/skimage.io.html#skimage.io.plugin_info
from skimage import data, io

imgSCIKIT = io.imread('.\\Imagenes\\lena_color_512.tif')
io.imshow(imgSCIKIT)
```

```python
#RGB a YUV
from skimage.color import rgb2yuv
yuv_img = rgb2yuv(imgSCIKIT)
io.imshow(yuv_img)
```

```python
#RGB a HSV
from skimage.color import rgb2hsv
hsv_img = rgb2hsv(imgSCIKIT)
io.imshow(hsv_img)
```

```python
#RGB a escala de grises
from skimage.color import rgb2grey
gray_img = rgb2grey(imgSCIKIT)
io.imshow(gray_img)
```
