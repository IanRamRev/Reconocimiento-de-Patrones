```python
#SCRIPT PARA ABRIR IMAGEN CON SCIKIT
#https://scikit-image.org/docs/stable/api/skimage.io.html#skimage.io.plugin_info
from skimage import data, io
from skimage.color import rgb2grey
import numpy as np
import pandas as pd

img = io.imread('.\\Imagenes\\lena_color_512.tif')
gray_img = rgb2grey(img)
io.imshow(gray_img)
print(gray_img)


def promediarLista(lista):
    sum=0.0
    for i in range(len(lista)):
        for j in range(len(lista[i])):
            sum=sum+lista[i][j]
    return sum/4
```

```python
#Tenemos 256 matrices de [2,512]
arraySegmentadoV = np.vsplit(gray_img, 256) #(0-255)
print("1 array con 256 subarrays de ",arraySegmentadoV[0].shape)
print (arraySegmentadoV)
print ("Arrays dentro de array principal: ", len(arraySegmentadoV))
```
