### Separación de canales

```python
imgLena = openImage(pathLena, 1)

imgR = imgLena.copy()
imgR[:,:,0] = 0 
imgR[:,:,1] = 0 

imgG = imgLena.copy()
imgG[:,:,0] = 0 
imgG[:,:,2] = 0 

imgB = imgLena.copy()
imgB[:,:,1] = 0 
imgB[:,:,2] = 0 


pl.figure(figsize=(12, 12), dpi=80)

plt.subplot(1, 4, 1)
plt.imshow(cv2.cvtColor(imgR, cv2.COLOR_BGR2RGB))
plt.title("Lena Rojo")
plt.axis('off')

plt.subplot(1, 4, 2)
plt.imshow(cv2.cvtColor(imgG, cv2.COLOR_BGR2RGB))
plt.title("Lena Verde")
plt.axis('off')

plt.subplot(1, 4, 3)
plt.imshow(cv2.cvtColor(imgB, cv2.COLOR_BGR2RGB))
plt.title("Lena Azul")
plt.axis('off')

plt.subplot(1, 4, 4)
plt.imshow(cv2.cvtColor(imgB, cv2.COLOR_BGR2RGB) + cv2.cvtColor(imgG, cv2.COLOR_BGR2RGB) + cv2.cvtColor(imgR, cv2.COLOR_BGR2RGB))
plt.title("Lena Azul")
plt.axis('off')
```

```python
imgPeppers = openImage(pathPeppers, 1)

imgR_Peppers = imgPeppers.copy()
imgR_Peppers[:,:,0] = 0 
imgR_Peppers[:,:,1] = 0 

imgG_Peppers = imgPeppers.copy()
imgG_Peppers[:,:,0] = 0 
imgG_Peppers[:,:,2] = 0 

imgB_Peppers = imgPeppers.copy()
imgB_Peppers[:,:,1] = 0 
imgB_Peppers[:,:,2] = 0 


pl.figure(figsize=(12, 12), dpi=80)

plt.subplot(1, 4, 1)
plt.imshow(cv2.cvtColor(imgR_Peppers, cv2.COLOR_BGR2RGB))
plt.title("Lena Rojo")
plt.axis('off')

plt.subplot(1, 4, 2)
plt.imshow(cv2.cvtColor(imgG_Peppers, cv2.COLOR_BGR2RGB))
plt.title("Lena Verde")
plt.axis('off')

plt.subplot(1, 4, 3)
plt.imshow(cv2.cvtColor(imgB_Peppers, cv2.COLOR_BGR2RGB))
plt.title("Lena Azul")
plt.axis('off')

plt.subplot(1, 4, 4)
plt.imshow(cv2.cvtColor(imgB_Peppers, cv2.COLOR_BGR2RGB) + cv2.cvtColor(imgG_Peppers, cv2.COLOR_BGR2RGB) + cv2.cvtColor(imgR_Peppers, cv2.COLOR_BGR2RGB))
plt.title("RGB")
plt.axis('off')
```
