```python
#SCRIPT PARA ABRIR IMAGEN CON PIL
# DOCUMENTACIÓN | https://pillow.readthedocs.io/en/4.2.x/reference/Image.html

import matplotlib.pyplot as plt
from PIL import Image

imagen = Image.open(".\\Imagenes\\lena_color_512.tif")
print('Formato:', imagen.format)
print('Tamaño:',imagen.size)
print('Información:', imagen.getdata)
print('Tipo de dato: ', type(imagen))
#plt.imshow(imagen,cmpap = "RGB")
plt.imshow(imagen)
```

```python
#SCRIPT PARA ABRIR IMAGEN CON OPENCV
import numpy as np
import matplotlib.pyplot as plt
import cv2

img = cv2.imread('.\\Imagenes\\lena_color_512.tif')
print('Formato:')
print('Tamaño:', img.size)
print('Tipo:', type(img))
print(img.dtype)
plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
```
