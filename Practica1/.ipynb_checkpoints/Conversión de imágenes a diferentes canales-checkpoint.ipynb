```python
import numpy as np
import matplotlib.pyplot as plt
from skimage import data, io
import cv2
%matplotlib inline



def openImage(path):
    img = cv2.imread(path)
    return img

def convertChannelOpenCV(img, channelType):
    if(channelType == 1):
        imgRGB = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
        return imgRGB
    if(channelType == 2):
        img_gray = cv2.cvtColor(img, cv2.COLOR_RGB2GRAY)
        return img_gray
    if(channelType == 3):
        img_yuv = cv2.cvtColor(img, cv2.COLOR_BGR2YUV)
        return img_yuv
    if(channelType == 4):
        img_hsv = cv2.cvtColor(img, cv2.COLOR_BGR2HSV)
        return img_hsv
        
        


#image pathÂ´s
pathLena = '.\\Imagenes\\lena_color_512.tif'
pathPeppers = '.\\Imagenes\\peppers_color.tif'
```

```python
#Convert RGB
io.imshow(convertChannelOpenCV(openImage(pathLena), 1))
```

```python
#Convert RGB
io.imshow(convertChannelOpenCV(openImage(pathPeppers), 1))
```

```python
#Convert RGB to gray scale
io.imshow(convertChannelOpenCV(openImage(pathLena), 2))
```

```python
#Convert RGB to gray scale
io.imshow(convertChannelOpenCV(openImage(pathPeppers), 2))
```

```python
#Convert RGB to YUV
io.imshow(convertChannelOpenCV(openImage(pathLena), 3))
```

```python
#Convert RGB to YUV
io.imshow(convertChannelOpenCV(openImage(pathPeppers), 3))
```

```python
#Convert RGB to HSV
io.imshow(convertChannelOpenCV(openImage(pathLena), 4))
```
