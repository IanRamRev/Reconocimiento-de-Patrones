```python
#SCRIPT PARA ABRIR IMAGEN CON PIL
# DOCUMENTACIÓN | https://pillow.readthedocs.io/en/4.2.x/reference/Image.html
# https://python-para-impacientes.blogspot.com/2014/12/fundamentos-para-procesar-imagenes-con.html

import matplotlib.pyplot as plt
from PIL import Image

imagen = Image.open(".\\Imagenes\\lena_color_512.tif")
print('Formato:', imagen.format)
print('Tamaño:',imagen.size)
print('Información:', imagen.getdata)
print('Tipo de dato: ', type(imagen))
plt.imshow(imagen)
```

```python
#SCRIPT PARA ABRIR IMAGEN CON MATPLOTLIB
import matplotlib.pyplot as plt
import matplotlib.image as mpimg

img=mpimg.imread('.\\Imagenes\\cameraman.tif')
imgplot = plt.imshow(img, cmap="gray")

print('Tamaño: ', img.shape[0:2])
print('Tipo de dato: ', type(img))

print(img)
```

```python
#SCRIPT PARA ABRIR IMAGEN CON OPENCV
#https://likegeeks.com/es/procesar-de-imagenes-en-python/

import numpy as np
import matplotlib.pyplot as plt
import cv2

# Load an color image in grayscale
img = cv2.imread('.\\Imagenes\\lena_color_512.tif',0)
dataImg = format(img)
print(dataImg)
print('Formato:')
print('Tamaño:', img.size)
print('Tipo:', type(img))
print(img.dtype)
imgplot = plt.imshow(img)
```

```python
#SCRIPT PARA ABRIR IMAGEN CON SCIKIT
#https://scikit-image.org/docs/stable/api/skimage.io.html#skimage.io.plugin_info
from skimage import data, io

img = io.imread('.\\Imagenes\\lena_color_512.tif')
#print('Datos: ', format(img))
print('Formato:')
print('Tamaño:', img.shape[0:2])
print('Tipo:', type(img))
io.imshow(img)
```
