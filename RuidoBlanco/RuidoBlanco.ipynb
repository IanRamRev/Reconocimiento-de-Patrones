```python
from skimage import data, io
from skimage.color import rgb2grey
import numpy as np
import matplotlib.pyplot as plt
import random

def creaMatrixRB(Matrix):
    for i in range(len(Matrix)):
        for j in range(len(Matrix[i])):
            Matrix[i][j] = random.random()
```

```python
matrixRB = np.zeros((256,256))
creaMatrixRB(matrixRB)
```

```python
transformacion = np.fft.fft2(matrixRB)
fshift = np.fft.fftshift(transformacion)
magnitudFFT = 20*np.log(np.abs(fshift))
plt.subplot(121),plt.imshow(matrixRB, cmap = 'gray')
plt.title('Imagen de entrada'), plt.xticks([]), plt.yticks([])
plt.subplot(122),plt.imshow(magnitudFFT, cmap = 'gray')
plt.title('FFT'), plt.xticks([]), plt.yticks([])
plt.show()
```
