```python
# Funcion para las mascaras

def mascar(tipo,ventana):
    if tipo==0:
        ventanagris=ventana*0+64 
    elif tipo==1:
        ventanagris=ventana*0+128
    elif tipo==2:
        ventanagris=ventana*0+192
    else:
        ventanagris=ventana*0+255
    return ventanagris
    
```
